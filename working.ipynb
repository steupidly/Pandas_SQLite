Load data into database

```python
import pandas as pd
import json
import numpy as np

pd.options.display.float_format= '{:.2f}'.format
```

```python
with open('some_movies.json') as f:
    data = json.load(f)
```

```python
df = pd.json_normalize(data, sep='_')
df.head()
```

```python
movies = df[["id", "title", "revenue", "budget", "belongs_to_collection_name", "release_date"]].copy()
movies
```

```python
movies.release_date = pd.to_datetime(movies.release_date)
```

```python
movies.revenue = movies.revenue.div(1000000)
movies.rename(columns={'revenue':'revenue_musd'}, inplace=True)
movies.head()
```

```python
movies.budget = movies.budget.div(1000000)
movies.rename(columns={'budget':'budget_musd'}, inplace=True)
movies
```

```python
votes = df[['id', 'vote_count','vote_average']].copy()
votes
```

```python
#normalize satu column json menjadi df baru 
genres = pd.json_normalize(data=data, record_path='genres',meta='id',record_prefix='genres_')
genres
```

```python
prod = pd.json_normalize(data=data, record_path='production_companies', meta='id', record_prefix='comp_')
prod
```

```python
con = sq3.connect('movies.db')
```

```python
con
```

```python
movies.to_sql('Movies',con, index=False)
```

```python
votes.to_sql('Votes',con, index=False)
```

```python
genres.to_sql('Genres',con, index=False)
```

```python
prod.to_sql('Production',con,index=False)
```

```python
con.execute("select * From sqlite_master").fetchall()
```

```python
con.execute("SELECT name FROM sqlite_master WHERE type='table' ORDER BY name").fetchall()
```

```python
con.close()
```

Join Query

```python
con=sq3.connect('movies.db')
```

```python
pd.read_sql('SELECT * FROM MOVIES\
            JOIN VOTES \
            ON MOVIES.ID = VOTES.ID', con)
```

```python
pd.read_sql('select m.id, title, v.vote_average from movies as m\
            join votes as v\
            on m.id = v.id order by m.id',con,index_col='id')
```

```python
pd.read_sql("select m.id, title, vote_average from movies as m\
            join votes as v \
            on m.id = v.id\
            where vote_average > 8 order by m.id  ",con, index_col= 'id')
```

```python
pd.read_sql('select m.id, title, vote_average, budget_musd from movies m\
            join votes v on m.id=v.id\
            where vote_average > 8 order by budget_musd asc',con,index_col='id')
```
